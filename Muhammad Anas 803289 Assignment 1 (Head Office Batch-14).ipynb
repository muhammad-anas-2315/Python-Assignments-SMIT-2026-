# 🔹 Q12 (Full Name + Upper/Lower)  User se first name aur last name lo aur:  1)  full name print karo  2) upper case 3)lower case

```python
first_name = input("Enter Your First Name:")
last_name = input("Enter Your Last Name:")
full_name = f"{first_name} {last_name}"
print(f"Full Name: {full_name}")
print(f"Upper Case: {full_name.upper()}")
print(f"Lower Case: {full_name.lower()}")
```

# 🔹 Q13 (Sentence Formatting) User se aik sentence lo aur usay: 1) title case 2)capitalize  3)length print karo

```python
sentence = input("Enter a sentence:")
print(f"Tittle Case: {sentence.title()}")
print(f"Capital Case: {sentence.capitalize()}")
print(f"Length: {len(sentence)}")
```

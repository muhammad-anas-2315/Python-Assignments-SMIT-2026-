# User se marks lo aur grade do:

# 90+ → A
# 80+ → B
# 70+ → C
# 60+ → D
# warna → Fail

```python
your_marks = float(input("Enter your Marks: "))
if your_marks >= 90:
    print("Grade: A")
elif your_marks >= 80:
    print("Grade: B")
elif your_marks >= 70:
    print("Grade: C")
elif your_marks >= 60:
    print("Grade: D")
else:
    print("Fail")
```

# User se day number (1–7) lo:

# 1 → Monday
# 2 → Tuesday
## ...
# 7 → Sunday

```python
number_day = int(input("Enter (1-7) Number for Days: "))
if number_day == 1:
    print("Monday")
elif number_day == 2:
    print("Tuesday")
elif number_day == 3:
    print("Wednesday")
elif number_day == 4:
    print("Thursday")
elif number_day == 5:
    print("Friday")
elif number_day == 6:
    print("Saturday")
elif number_day ==7:
    print("Sunday")
else:
    print("invalid error")
```

# User se month number lo:

# 12,1,2 → Winter
# 3,4,5 → Spring
# 6,7,8 → Summer
# warna → Autumn

```python
month_num = int(input("Enter Number (1-12) of Month"))
if month_num == 12 or month_num == 1 or month_num == 2:
    print("Winter")
elif month_num == 3 or month_num == 4 or month_num == 5:
    print("Spring")
elif month_num == 6 or month_num == 7 or month_num == 8:
    print("Summer")
else:
    print("Autumn")
```

# User se 2 numbers lo:

# pehla > dusra → "First is greater"
# pehla < dusra → "Second is greater"
# warna → "Both equal"

```python
user_num1 = float(input("Enter 1st Number: "))
user_num2 = float(input("Enter 2nd Number: "))
if user_num1 > user_num2:
    print("First is Greater")
elif user_num1 < user_num2:
    print("Second is Greater")
else:
    print("Both Equal")
```

# Simple calculator:
# User se operator lo (+, -, *, /):

# → addition
# → subtraction
# → multiplication
# / → division
# warna → "Invalid operator"

```python
operator = input("Enter Operator Sign (+, -, *, /): ")
if operator == "+":
    print("addition")
elif operator == "-":
    print("subtraction")
elif operator == "*":
    print("multiplication")
elif operator == "/":
    print ("division")
else:
    print("invalid operator")
```
